Converting text into vectors

```python
import os
from dotenv import load_dotenv
load_dotenv() # load all the environment variables
```

```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPEN_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model = "text-embedding-3-large")
embeddings
```

```python
text = "Generative AI is awesome!"
query_result = embeddings.embed_query(text) # Takes the sentence and converts it into a vector of dimension: 3072 (dimension specified in documentation)
query_result
```

```python
len(query_result)
```

```python
# can specify a different dimension also
embeddings_1024 = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)
embeddings_1024
```

```python
text = "Generative AI is awesome!"
query_result = embeddings_1024.embed_query(text) 
len(query_result)
```

```python
query_result
```

```python
# Load document using TextLoader

from langchain_community.document_loaders import TextLoader
loader = TextLoader("speech.txt")
docs = loader.load()
docs
```

```python
# Split the document using Recursive Text Splitter

from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_documents=text_splitter.split_documents(docs)
final_documents
```

```python
final_documents[0:3]
```

```python
# Next, we need to take all these documents, convert these into vectors and store them in a vector db.

## combine vector embedding and vector store db - use chroma db (open source)
from langchain_community.vectorstores import Chroma
db = Chroma.from_documents(final_documents[0:3], embeddings_1024)
# apply embedding technique given as param 2 to all final_documents given as param 1.
db
```

```python
# how to query from vectorstore database

query = "It will be all the easier for us to conduct ourselves as belligerents"
retrieved_results = db.similar_search(query)
print(retrieved_results)
```
