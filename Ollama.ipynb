OLLAMA EMBEDDINGS

Download from https://ollama.com/download/windows and install - will show up in hidden icons.

- See all supported models here: https://github.com/ollama/ollama

Ollama supports embedding model, making it possible to build retrieval augmented generation (RAG) application that combine text prompts with existing documents or other data.


```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = (
    OllamaEmbeddings(model = "gemma:2b") ## if model not given in params, by default it uses llama 2
)
```

```python
embeddings
```

```python
r1 = embeddings.embed_documents(
    [
        "Alpha is the first letter of Greek alphabet",
        "Beta is the second letter of Greek alphabet",
    ]
)
```

```python
r1
```

```python
len(r1[0]), len(r1[1])
```

```python
### Other embedding models
### https://ollama.com/blog/embedding-models
embeddings = OllamaEmbeddings(model = "mxbai-embed-large")
text = "This is a text document."
query_result = embeddings.embed_query(text)
query_result
```
